```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Load the Data Frame

file = r'D:\NIKHILESH\ML_AI\Data Handling with Python\bank-full.csv'
df = pd.read_csv(filepath_or_buffer = file,delimiter = ';')
df
```

```python
Density_plot_Age = sns.displot(df['age'],bins =(df['age'].max()-df['age'].min()))
# bins are number or bar. between range.
```

```python
#help(sns.countplot)
```

```python
#save figer.
Density_plot_Age.savefig('Density_plot_Age.png')
```

```python
# kde curve plot
sns.kdeplot(df['age']) #  kernel density estimation,
#a method for visualizing the distribution of observations in a dataset, analogous to a histogram.
```

```python
# box plot
sns.boxplot(x = df['age'])
```

```python
sns.boxplot(x = df['age'], y = df['marital'])
```

```python
#join plot
sns.jointplot(data = df.loc[(df['balance']>0) & (df['balance']<1500)], y = 'balance' ,x = 'age', kind = 'kde')
# x,y plot with count/frequence of x,y element on side graph.
```

```python
# liner model plot
sns.lmplot(data = df.loc[(df['duration']>0) & (df['duration']<400)], y = 'duration' ,x = 'age',order = 3)
```

```python
sns.lmplot(data = df.iloc[1:500,:], x = 'duration' ,y = 'campaign', hue = 'housing', col = 'default' , row = 'loan', order = 1)
# hue : The hue parameter in seaborn histograms allows you to color the bars in the histogram based on the value of some
#specific variable.This can be useful for comparing the distribution of a variable across multiple groups. 
#Like yes and no are colour diffrently.

# col will divid the polt in to n number of plot like in default there are two 'yes' and 'no' side by side
# row same like col but in up and down. 
```

```python
# count plot 
sns.countplot(data = df.iloc[1:500,:], x = 'education' , hue = 'housing')
```
